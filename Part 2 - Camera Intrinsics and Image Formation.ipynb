## Part 2 - Camera Intrinsics and Image Formation

```python
%matplotlib widget

import matplotlib.pyplot as plt
from utils import *

# In this example, we’ll consider a simple setup where 
# the camera is located at the origin and the image plane 
# is above it along the +ve Z-axis (We’ll be working with 
# a left-handed coordinate system). Next, we plot some points 
# such that all of them lie on the same plane parallel to 
# the image plane and top of it. This way it’s easy to 
# visualize the projections and image formation. Next, we 
# create the camera intrinsic matrix and use it to project 
# the points onto the image plane and form the image. Finally, 
# we transform the image plane using this matrix.
```

```python
# define parameters
# points
n_points = 6
xlim = (-5, 5)
ylim = (-5, 5)
elevation = 5

# image plane
img_size = (7, 7)
f = 2
```

```python
# We create 6x6 number of points on the same plane parallell with the image plane

points = create_same_plane_points(n_points, xlim, ylim, elevation)
xx, yy, Z = create_image_grid(f, img_size)
```

```python
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111,projection='3d')

# plot image plane
ax.plot_surface(xx, yy, Z, alpha=0.75)

ax.set(xlim=(-10, 10), ylim=(-10, 10), zlim=(0, 10))
ax = pr.plot_basis(ax)

# plot the points
c = 0
for i in range(n_points):
    for j in range(n_points):
        point = points[:, c]
        ax.scatter(*point, color="orange")
        ax.plot(*make_line(origin, point), color="purple", alpha=0.25)
        c += 1
        
ax.set_xlabel("X-axis")
ax.set_ylabel("Y-axis")
ax.set_zlabel("Z-axis")
```

### Camera Intrinsics and Image formation

```python
# Four different modes of the camera intrinsics.

types = ["ideal", "skewed", "rectangular pixels", "offset"]
K1 = compute_intrinsic_parameter_matrix(2, 0, 1, 0, 0) # normal / ideal
K2 = compute_intrinsic_parameter_matrix(2, 1, 1, 0, 0) # skewed
K3 = compute_intrinsic_parameter_matrix(2, 0, 2, 0, 0) # rectangular pixels
K4 = compute_intrinsic_parameter_matrix(2, 0, 1, 1, 1) # offset

projection_points1 = compute_image_projection(points, K1)
projection_points2 = compute_image_projection(points, K2)
projection_points3 = compute_image_projection(points, K3)
projection_points4 = compute_image_projection(points, K4)
```

```python
# Here the projections (images) from the 4 different intrincs are shown.

h, w = img_size
nrows = 2
ncols = 2

fig, axes = plt.subplots(nrows=nrows, ncols=ncols, figsize=(6, 4))

c = 1
for i in range(ncols):
    for j in range(nrows):
        ax = axes[i][j]
        ax.set(xlim = (-(h // 2), w // 2), ylim = (-(h // 2), w // 2))
        ax.set_title(types[c-1])
        for k in range(n_points * n_points):
                ax.scatter(*eval(f"projection_points{c}")[:, k])
        c += 1

plt.tight_layout()
```

### Transforming the image plane

```python
# To see the images the way the camera with certain intrinsics see it, we need to transform the image
# plane. The intrinsic matrix is a "change of basis"-matrix. If we take the inverse of a "change of basis"
# matrix, it becomes a transformation matrix. In order to see the image plane with the different intrinsic 
# camera paramters, we need the transformation matrix, i.e the inverse of the intrinsic matrix. 

# 1. extract image plane points
h_grid = convert_grid_to_homogeneous(xx, yy, Z, img_size)
h_grid = h_grid[:-1, :] # remove last row as we do not need homogeneous points

# 2. transform the image plane points
K_ = K4.copy() # Here we decide which intrinsic type we want
#K_[:2, :2] = 1/f * K_[:2, :2] # nullify the effect of focal length
Kinv = np.linalg.inv(K_) # the transformation matrix is the inverse of the change of basis matrix
img_plane_points_transformed = Kinv @ h_grid

xxt, yyt, Zt = convert_homogeneous_to_grid(img_plane_points_transformed, img_size)
```

```python
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111,projection='3d')

# plot image plane
ax.plot_surface(xxt, yyt, Zt, alpha=0.75)

ax.set(xlim=(-10, 10), ylim=(-10, 10), zlim=(0, 10))
ax = pr.plot_basis(ax)

# plot the points
c = 0
for i in range(n_points):
    for j in range(n_points):
        point = points[:, c]
        ax.scatter(*point, color="orange")
        ax.plot(*make_line(origin, point), color="purple", alpha=0.25)
        c += 1
        
ax.set_xlabel("X-axis")
ax.set_ylabel("Y-axis")
ax.set_zlabel("Z-axis")
```
